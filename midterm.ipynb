```python
##sum up 1/n for n=1 to 10000
def sum_up(n):
    sum = 0
    for i in range(1,n+1):
        sum += 1/i
    return sum

#round each number 1/n to 5 decimals then sums them up in 5-digit decimal arithmetic for n=1 to 10000
def sum_up_rounded(n):
    sum = 0
    for i in range(1,n+1):
        sum += round(1/i,5)
    return sum

#sums same rounded numbers but in reverse order for n=10000 to 1
def sum_down_rounded(n):
    sum = 0
    for i in range(n,0,-1):
        sum += round(1/i,5)
    return sum


print(f'(a): {sum_up(10000)}\n(b): {sum_up_rounded(10000)}\n(c): {sum_down_rounded(10000)}')
```

From this output we can denote round-off errors. Naturally, it follows that in the unrounded summation, (a), it yields the closest result to the true value for the harmonic series $\frac{1}{n}$. When comared to the other two results, we can see that (a) is slightly larger. More interestingly, we can see that the rounded summations in (b) and (c), there is a very slight difference in the result. This slight difference comes from the order of operations that each algorithim performs its computations. This is likely due to the fact that floating-point arithemtic is NOT assosciative. That is:
$$
a + b \neq b + a
$$
Thus, the order in which we add our numbers affects our final result. In the forward sum, we add the largest numbers ($\frac{1}{n}$ becomes smaller as $n$ grows large) our result starts to grow larger; and as our terms become smaller (larger $n$) they start to contribute less and less to the final result due to loss of precision in rounding. In the case of the backward sum, adding the smallest numbers while our total sum is stil small relatively maintains those small values's precision and thus as a result contribute more to the final result.

Now we will show that the convergence rate is not quadratic and similar to thar of the the bisection method (linear). We say a method is linearly convergent if $\exists$ a constant $\rho$ such that
$$ 
|x_{k+1} - x^*| \leq \rho |x_k - x^*|

$$

From earlier, we showed that $x^*$ is a double root for $x=1$.

Let $e_k = x_k - x^* \implies x_k = e_k + x^*$

$$

$$

We have found our $\rho = \frac{1}{2}$, implying that for this method our error at each step is half the error at the previous step demonstrating linear convergence. 

```python
#Implementation of Newton's method starting from x_0 = 2
import math
import matplotlib.pyplot as plt
import numpy as np

def newton_method(f, f_prime, x_0, tol, max_iter=50):
    iterates = [x_0]
    iter = 0
    error = float('inf')

    while error > tol and iter < max_iter:
        x_1 = x_0 - f(x_0)/f_prime(x_0)

        error = abs(x_1 - x_0)
        print(f'Iteration k={iter+1}: x_{iter+1} = {x_1}. Error: {error}')

        iterates.append(x_1)
        x_0 = x_1
        iter += 1
    return iterates

f = lambda x: math.exp(x) * (x-1)**2
x_0 = 2
f_prime = lambda x: math.exp(x)*((x**2) - 1 )

# Apply Newton's method with a tolerance level of 1e-10
iterates = newton_method(f, f_prime, x_0, 1e-10)


```

Given the 2nd order non-linear ordinary differential equation
$$
\dfrac{d^2y}{dx^2} = - \lparen\dfrac{dy}{dx}\rparen^2 - y  + \ln x, \qquad \text{for} \ 1 \leq x \leq 2, \qquad \text{with initial conditions} \  y(1) = 0 , \ y(2) = \ln 2
$$
We define its discretization as
$$
\frac{y_{i+1} - 2y_i + y_{i-1}}{h^2} + (\frac{y_{i+1} - y_{i-1}}{2h})^2 + y_i = \ln(ih), \qquad i = 1,2,\dots,n
$$
So
$$
f_i(x) = \frac{y_{i+1} - 2y_i + y_{i-1}}{h^2} + (\frac{y_{i+1} - y_{i-1}}{2h})^2 + y_i - \ln(ih),
$$
Then we can construct our sysem of equations $\vec{F}$

$$
\vec{F}(y) = 
\begin{bmatrix}
f_0(y) \\
f_1(y) \\
f_2(y) \\
\vdots \\
f_{n-1}(y) \\
f_n(y) \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
\vdots \\
0 \\
0 \\
\end{bmatrix}
$$

For our Jacobian
$$
J_{ij} = \dfrac{\partial f_i}{\partial y_j}
$$

The derivative of our equation with respect to $y_i$ is
$$
\dfrac{\partial f_i}{\partial y_i} = 1 -\frac{2}{h^2}
$$

The derivative of our equation with respect to $y_{i+1}$ is
$$
\dfrac{\partial f_{i}}{\partial y_{i+1}} = \frac{1}{h^2} + \frac{y_{i+1}-y_{i-1}}{2h^2} = \frac{2 + y_{i+1}-y_{i-1}}{2h^2}
$$

The derivative of our equation with respect to $y_{i-1}$ is
$$
\dfrac{\partial f_{i}}{\partial y_{i-1}} = \frac{1}{h^2}  - \frac{y_{i+1}-y_{i-1}}{2h^2} = \frac{2 - y_{i+1} + y_{i-1}}{2h^2}
$$

In the case of our diagonal, or when $i=j$, it follows that
$$
J_{ii} = 1-\frac{2}{h^2} 
$$
And for $i = j + 1 $, our lower diagonal
$$
J_{i,i-1} = \frac{2 - y_{i+1} + y_{i-1}}{2h^2}
$$
Finally, for our upper diagonal $i = j - 1$
$$
J_{i,i+1} =  \frac{2 + y_{i+1}-y_{i-1}}{2h^2}
$$

From this, it is evident that our Jacobian will be tridiagonal with elements off the diagonal dependent on $y_i$ and the off-diagonals dependent on $y_{i+1}$ and $y_{i-1}$ \
\
Thus we obtain our Jacobian

$$
J = \begin{bmatrix}
1-\frac{2}{h^2} & \frac{2 + y_2 - y_0}{2h^2} & 0 & \cdots & 0 \\
\frac{2 - y_3 - y_1}{2h^2} & 1-\frac{2}{h^2} & \frac{2 + y_3 - y_1}{2h^2} & \cdots & 0 \\
0 & \frac{2 - y_4 - y_2}{2h^2} & 1-\frac{2}{h^2} & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \cdots & 1-\frac{2}{h^2}
\end{bmatrix}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to represent the discretized system of equations
def discretized_system(y, n, h):
    equations = np.zeros(n+1)
    equations[0] = y[0] - 0 # y(1) = 0
    equations[n] = y[n] - np.log(2) # y(2) = ln(2)
    for i in range(1, n):
        xi = (i)*h + 1
        fi = ((y[i+1] - 2*y[i] + y[i-1])/h**2 +
              ((y[i+1] - y[i-1])/(2*h))**2 +
              y[i] - np.log(xi))
        equations[i] = fi
    return equations #system of equations F(y) = 0

# Construct the Jacobian matrix
def jacobian(y, h, n):
    J = np.zeros((n+1, n+1))
    np.fill_diagonal(J, 1 - 2/h**2)
    for i in range(1, n):
        J[i, i+1] = (2 + y[i+1] - y[i-1]) / (2 * h**2)
        J[i, i-1] = (2 - y[i+1] + y[i-1]) / (2 * h**2)
    J[0, 0] = J[n, n] = 1  # y(1) = 0 and y(n) = ln(2) are fixed, so derivative is 1 and no other dependencies
    return J

# Newton's method for solving ODE
def newtons_method(n):
    h = 1/n
    avg_boundary = (0 + np.log(2)) / 2
    y = np.full(n+1, avg_boundary)
    max_iter = 100 
    tol = 1e-10 

    for iteration in range(max_iter):
        f_val = discretized_system(y, n, h)
        J = jacobian(y, h, n)
        
        p = np.linalg.solve(J, -f_val)
        y += p  # update the guess
        
        error = np.linalg.norm(p)
        print(f'Iteration {iteration}:, Change p {p}, Error = {error}')
        
        if error < tol:
            print(f'Convergence achieved after {iteration} iterations.')
            break
    
    return y

# Mesh sizes to use
mesh_sizes = [8, 16, 32, 64]

# Solve for each mesh size and print the solution
for n in mesh_sizes:
    print(f'\nSolving for n={n} with Newton\'s method...')
    solution = newtons_method(n)
    x_vals = np.linspace(1, 2, n+1)
    actual_solution = np.log(x_vals)
    error = np.abs(solution - actual_solution)

    # Plot the numerical solution vs the actual solution
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, solution, 'o-', label=f'Numerical Solution (n={n})')
    plt.plot(x_vals, actual_solution, 'x-', label='Actual Solution (ln(x))')
    plt.title(f'Numerical vs Actual Solution for n={n}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print out the maximum error
    max_error = np.max(error)
    print(f'Maximum error for n={n}: {max_error}')

    # Check if the solution satisfies the boundary conditions
    bc_error1 = np.abs(solution[0] - 0)
    bc_error2 = np.abs(solution[-1] - np.log(2))
    print(f'Boundary condition error at y(1) for n={n}: {bc_error1}')
    print(f'Boundary condition error at y(2) for n={n}: {bc_error2}')
```

Newton's method shows  efficiency in our case, achieving convergence within just 4 iterations across all mesh sizes, indicating that our initial guess is particularly effective. This is likely due to the small interval ([1, 2]) of the problem, where a simple linear interpolation between the boundary conditions seems to suffice as a good starting point. Furthermore, an important observation is the decrease in maximum error with increasing mesh sizes, affirming that the accuracy of our solution improves with finer discretization. This trend is typical of well-behaved numerical methods for solving ODEs and suggests that our discretization strategy is good.
